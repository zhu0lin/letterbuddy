```python
#temporarily outdated -  /results_paired_Human structure changing in next commit


import os

def accuracyTester(answers="./results_paired_Human", attempted="./results_paired_EMNIST"):
    """
    Compare human-labeled images with model-predicted folders.
    
    - Human files: char_102N.png (last character = ground truth)
    - EMNIST files: attempted/<predicted_char>/char_102.png
    Logic:
      1. For each human file, extract base and human char.
      2. Search all folders in attempted.
      3. For each folder, search files; if base matches, compare folder vs human char.
      4. Track GOOD / BAD / NOT FOUND.
    """
    total = 0
    correct = 0
    not_found = 0

    for fname in os.listdir(answers):
        if not fname.lower().endswith((".png", ".jpg", ".jpeg")):
            continue

        total += 1
        fChar = fname[-5] if fname.endswith(".png") else fname[-4]  # last char before extension
        base = fname[:-5] if fname.endswith(".png") else fname[:-4]  # remove last char + extension

        found_match = False

        # Search each folder in attempted
        for charFolder in os.listdir(attempted):
            folder_path = os.path.join(attempted, charFolder)
            if not os.path.isdir(folder_path):
                continue

            for file in os.listdir(folder_path):
                if os.path.splitext(file)[0] == base:  # match base name
                    found_match = True
                    charThought = charFolder
                    if charThought == fChar:
                        correct += 1
                        print(f"Comparing {base}, human: '{fChar}', predicted folder: '{charThought}'")
                        print("✅ GOOD")
                    else:
                        print(f"Comparing {base}, human: '{fChar}', predicted folder: '{charThought}'")
                        print("❌ BAD (folder doesn't match human label)")
                    break  # stop after first match
            if found_match:
                break

        if not found_match:
            not_found += 1
            print(f"⚠️ No prediction found for {base}, human: '{fChar}'")

    if total == 0:
        accuracy = 0.0
    else:
        accuracy = correct / total

    print(f"\nTotal: {total}, Correct: {correct}, Not found: {not_found}")
    print(f"Overall accuracy: {accuracy*100:.2f}%")
    return accuracy


# import os

# def accuracyTester(answers="./results_paired_Human", attempted="./results_paired_EMNIST"):
#     """
#     Compares human-labeled results with model-predicted results.
#     Accuracy = (correct matches) / (total files).
#     """
#     total = 0
#     correct = 0

#     # loop through all human-labeled files
#     for fname in os.listdir(answers):
#         if not fname.endswith(".png"):  
#             continue

#         # ground truth is last character before extension
#         truth = fname[-5]   # "char_12g.png" -> "g"

#         # prediction folder structure = attempted/[predicted_char]/char_12.png
#         for pred_char in os.listdir(attempted):
#             pred_folder = os.path.join(attempted, pred_char)
#             pred_path = os.path.join(pred_folder, fname[:-5] + ".png")  # match char_12.png

#             if os.path.exists(pred_path):
#                 total += 1
#                 if pred_char == truth:
#                     correct += 1
#                 break  # found the file, no need to check other folders

#     if total == 0:
#         return 0.0
    
#     return correct / total

    
```
